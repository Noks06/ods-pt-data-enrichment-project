```python

import pandas as pd
import numpy as np
```

```python

basics_url="https://datasets.imdbws.com/title.basics.tsv.gz"
akas_url="https://datasets.imdbws.com/title.akas.tsv.gz"


#df_basic = pd.read_csv(basics_url)
df_basics = pd.read_csv(basics_url, sep='\t', low_memory=False)

```

```python
df_basics.tail(5)
```

```python
df_basics = df_basics.replace({'\\N':np.nan})
df_basics.tail()
```

```python
# basic info
df_basics.info()
```

```python
# Akas dataframe 
df_akas =pd.read_csv(akas_url, sep='\t', low_memory=False)
df_akas.head()
```

```python
df_akas= df_akas.loc[df_akas['region'] == 'US']
df_akas
```

```python
df_akas = df_akas.replace({'\\N':np.nan})
df_basics.head()
```

```python
#filter out movies that runtime column that is  null   
df_basics = df_basics.dropna(subset='runtimeMinutes')
df_basics['runtimeMinutes'].isna().sum()
```

```python
# filter movies that null for genra
df_basics = df_basics.dropna(subset='genres')
```

```python
df_basics.shape
```

```python
# keep startYear 2000-2022
df_basics = df_basics.where(df_basics['startYear'] >='2000')
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
keepers =df_basics['tconst'].isin(df_akas['titleId'])
keepers
df_basics = df_basics[keepers]
df_basics
```

```python
# Eliminate movies that include "Documentary" in genre (see tip below)
# Exclude movies that are included in the documentary category.z
is_documentary = df_basics['genres'].str.contains('documentary', case = False,na=False)
df_basics = df_basics[~is_documentary]

df_basics.head()
```

```python
# Keep only US movies (Use AKAs table, see "Filtering one dataframe based on another" section below)
```

```python
# keep titleType == movie

df_basics = df_basics.loc[df_basics['titleType'] == 'movie']
df_basics
```

```python
df_basics
```

```python
rating_url="https://datasets.imdbws.com/title.ratings.tsv.gz"
# dataframe for rating
df_rating =pd.read_csv(rating_url, sep='\t', low_memory=False)
df_rating
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
ratingkeepers =df_rating['tconst'].isin(df_akas['titleId'])

df_rating = df_rating[ratingkeepers]
df_rating

```

```python
# Replacing '\N ' with NAN 

df_rating = df_rating.replace({'\\N':np.nan})
df_rating.tail()
```

```python
# example making new folder with os

import os
os.makedirs('Data/',exist_ok=True) 
# Confirm folder created
os.listdir("Data/")
```

```python
## Save current dataframe to file.
df_rating.to_csv("Data/title_rating.csv.gz",compression='gzip',index=False)
df_akas.to_csv("Data/title_akas.csv.gz",compression='gzip',index=False)
df_basics.to_csv("Data/title_basics.csv.gz",compression='gzip',index=False)
```
